```python
#import dataset using pandans
import pandas as pd
data = pd.read_csv(r"C:\Users\Lenovo\Dropbox\PC\Desktop\Dataset.csv")
print(data)
```

```python
#Data Cleaning
#1. Remove the rows which has null values
data.dropna(inplace = True)
print(data)
```

```python
#2. Correct the format in the given data
data["YearBuilt"] = data["YearBuilt"].astype('int')
data['Date'] = pd.to_datetime(data['Date'])
print(data)
```

```python
#3. Given that there are no hoses having more than 6 rooms. So, removing rows which has more than 6 rooms:
for i in data.index:
    if data.loc[i, "Rooms"]>6:
        data.drop(i, inplace = True)
print(data)
```

```python
#4. Remove the duplicate values
data.drop_duplicates(inplace = True)
print(data)
```

```python
#5. Remving rows which has 0 as size of the land
data = data[data["Landsize"] > 0]
data
```

```python
#6 No. of cars cannot be more than . If it is more than 6 cars, it is regarded as wrong entry
for i in data.index:
    if data.loc[i, "Car"]>6:
        data.drop(i, inplace = True)
print(data)
```
